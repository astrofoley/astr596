```python
# con = sq3.connect('../data/kaepora_foundation_only_v2.db')
# cur = con.cursor()
# con.execute("""DROP TABLE IF EXISTS Spectra""")
# con.execute("""CREATE TABLE IF NOT EXISTS Spectra (Filename TEXT PRIMARY KEY, SN Text, Source Text, Phase REAL, MinWave REAL, MaxWave REAL, snr REAL, Interpolated_Spectra BLOB, MJD REAL, Ref Text)""")
num_added = 0
files_not_added = []
# for i, spec_file in enumerate(files):
for i, spec_file in enumerate(files+files_non_team):
    with open(spec_file) as spec:
        filename = spec_file.split('/')[-1]
        if 'foundation' in filename.lower():
            spectrum = np.genfromtxt(spec_file)
            min_wave = spectrum[0][0]
            max_wave = spectrum[-1][0]
        else:
            spectrum = np.genfromtxt(spec_file, skip_header=9)
            min_wave = spectrum[0][0]
            max_wave = spectrum[-1][0]
            
        spec_meta = spec.readlines()[0:6]
        sn_name = spec_meta[0].split('#')[1].split()[-1].strip().lower()
        if '-' in sn_name:
            sn_name = sn_name.replace('-', '')
        date = spec_meta[1].split('#')[1].split()[-1].strip().lower()
        kyle_peak = spec_meta[3].split('#')[1].split()[-1].strip().lower()
        kyle_z = spec_meta[4].split('#')[1].split()[-1].strip().lower()
        kyle_phase = spec_meta[5].split('#')[1].split()[-1].strip().lower()
        
        if 'foundation' in filename.lower() or 'nonteam' in filename.lower():
            newdate = date[0:4] + '-' + date[4:6] + '-' + date[6:] + 'T00:00:00'
            mjd = Time(newdate, format='isot', scale='utc').mjd
        else:
            mjd = float(date)
            
        if sn_name in names and sn_name not in more_names and sn_name in mlcs_tmaxes.keys():
#             mjd_max = all_meta['t0'][names.index(sn_name)]
#             redshift = all_meta['z'][names.index(sn_name)]
            mjd_max = mlcs_tmaxes[sn_name]
            redshift = mlcs_redshifts[sn_name]
            phase = (mjd - mjd_max)/(1.+redshift)
        elif sn_name in more_names and sn_name in mlcs_tmaxes.keys():
            mjd_max = mlcs_tmaxes[sn_name]
            redshift = mlcs_redshifts[sn_name]
            phase = (mjd - mjd_max)/(1.+redshift)
        else:
            mjd_max = None 
            redshift = None 
            phase = None 
        
        if phase == 'nan' or mjd_max == 'nan':
            phase = None
            mjd_max = None

        
        source = 'foundation'
        ref = 'Dettman et al. 2020'
        if redshift != None:
            num_added+=1
            
            if np.absolute(float(kyle_phase) - phase) > 2:
                print sn_name, float(kyle_phase) - phase
                print kyle_z, redshift
                print kyle_peak, mjd_max
                print kyle_phase, phase
                print
#             print sn_name, filename, mjd, mjd_max, phase, redshift, min_wave, max_wave
            interp_spec, sig_noise = prep.compprep(spectrum, sn_name, redshift, source, testing = True, filename=filename, mjd=mjd, mjd_max=mjd_max)
            interped = msg.packb(interp_spec)
#             con.execute("""INSERT INTO Spectra(Filename, SN, Source, Phase, MinWave, MaxWave, snr, Interpolated_Spectra, MJD, Ref) VALUES(?, ?, ?, ?, ?, ?, ?, ?, ?, ?)""",
#                         (filename, sn_name, source, phase, min_wave, max_wave,
#                          sig_noise, buffer(interped), mjd, ref))
        else:
            files_not_added.append(filename)
print num_added
con.commit()
```
